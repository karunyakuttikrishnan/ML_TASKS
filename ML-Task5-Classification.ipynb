# CLASSIFICATION & CLUSTERING ALGORITHMS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Reading the dataset

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
x=iris.data
y=iris.target
```

## 1. LOGISTIC REGRESSION

### Splitting the dataset, training and testing

```python
from sklearn.model_selection import train_test_split
LX_train,LX_test, Ly_train, Ly_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
```

```python
lg.fit(LX_train, Ly_train)
```

### Model Evaluation

##### Confusion Matrix

```python
y_predicted1 = lg.predict(LX_test)
```

```python
from sklearn.metrics import confusion_matrix
cm1 = confusion_matrix(Ly_test, y_predicted1)
cm1
```

```python
import seaborn as sn
plt.figure(figsize = (5,4))
sn.heatmap(cm1, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
TP = np.diag(cm1)

# True Negatives (TN)
TN = np.sum(cm1) - (np.sum(cm1, axis=0) + np.sum(cm1, axis=1) - TP)

# False Positives (FP)
FP = np.sum(cm1, axis=0) - TP

# False Negatives (FN)
FN = np.sum(cm1, axis=1) - TP

# Accuracy
accuracy = (TP + TN) / np.sum(cm1)

# Precision
precision = TP / (TP + FP)

# Recall
recall = TP / (TP + FN)

# Specificity
specificity = TN / (TN + FP)

# F1 Score
f1_score = 2 * (precision * recall) / (precision + recall)

# False Positive Rate
false_positive_rate = FP / (FP + TN)

# False Negative Rate
false_negative_rate = FN / (FN + TP)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("Specificity:", specificity)
print("F1 Score:", f1_score)
print("False Positive Rate:", false_positive_rate)
print("False Negative Rate:", false_negative_rate)
```

##### Accuracy

```python
from sklearn.metrics import accuracy_score
Laccuracy = accuracy_score(Ly_test, y_predicted1)
print("Accuracy:", Laccuracy)
```

##### Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(y_predicted1,Ly_test))
```

##### Accuracy value is close to 1 which implies it is an ideal model. 
##### Precision value is to 1 indicating most true predictions are correct
##### Recall value is close to 1 indicating most positives are correctly predicted
##### Specificity value is close to 1 indicating most negatives are correctly predicted
##### F1 score value is close 1 indicating a perfect balance between precision and recall
##### FP rate values are close to 0 indicating only a very few negatives are falsely identified as positive
##### FN rate values are close to 0 indicating only a few positives are falsely identified as negative

### 6. K-MEANS CLUSTERING

### a) Elbow method 

```python
from sklearn.cluster import KMeans

k_range = range(1, 10)
sse = [] #sum of squared Error

for k in k_range:
    km = KMeans(n_clusters = k)
    km.fit(x)
    sse.append(km.inertia_) #this function is used to calculate the sum of square error
sse
```

```python
plt.xlabel("K value")
plt.ylabel("SSE")
plt.plot(k_range, sse)
```

##### From the above plot we observe that ideal value of k is 3 for our classification

### Training and testing the model

```python
km = KMeans(n_clusters=3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 25)
y_predicted6 = km.fit_predict(x)
y_predicted6
```

```python
#Checking centroid of the model
km.cluster_centers_
```

```python
# Visualising the clusters - On the first two columns
plt.scatter(x[y_predicted6 == 0, 2], x[y_predicted6 == 0, 3], s = 100, c = 'red', label = 'cluster 1')
plt.scatter(x[y_predicted6 == 1, 2], x[y_predicted6 == 1, 3], s = 100, c = 'blue', label = 'cluster 2')
plt.scatter(x[y_predicted6 == 2, 2], x[y_predicted6 == 2, 3], s = 100, c = 'green', label = 'cluster 3')

# Plotting the centroids of the clusters
plt.scatter(km.cluster_centers_[:, 2], km.cluster_centers_[:,3], s = 400, marker="*" ,c = 'black', label = 'Centroids')

plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.legend()
```

### b) Silhouette Method

```python
from sklearn.metrics import silhouette_score
#from sklearn.datasets import make_blobs

range_n_clusters = range(2, 10)
silhouette_avg_scores = []

for n_clusters in range_n_clusters:
    # Initialize KMeans with the current number of clusters
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(x)
    
    # Calculate the average silhouette score
    silhouette_avg = silhouette_score(x, cluster_labels)
    silhouette_avg_scores.append(silhouette_avg)
    print(f"For n_clusters = {n_clusters}, the average silhouette score is {silhouette_avg:.4f}")

# Plot silhouette scores for different values of k
plt.figure()
plt.plot(range_n_clusters, silhouette_avg_scores, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Average Silhouette Score')
plt.title('Silhouette Analysis for K-means Clustering')
plt.show()
```

##### Highest value of silhouette score is obtained when k=2.

### Training and testing the model

```python
skm = KMeans(n_clusters=2, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 25)
y_predicted6 = skm.fit_predict(x)
y_predicted6
```

```python
#Checking centroid of the model
skm.cluster_centers_
```

```python
# Visualising the clusters - On the first two columns
plt.scatter(x[y_predicted6 == 0, 0], x[y_predicted6 == 0, 1], s = 100, c = 'red', label = 'cluster 1')
plt.scatter(x[y_predicted6 == 1, 0], x[y_predicted6 == 1, 1], s = 100, c = 'blue', label = 'cluster 2')

# Plotting the centroids of the clusters
plt.scatter(skm.cluster_centers_[:, 0], skm.cluster_centers_[:,1], s = 400, marker="*" ,c = 'black', label = 'Centroids')

plt.legend()
```
